```python
def concat(a,b):
    return str(a) + "-" + str(b)

#axis=1 executes by row
# lambda means run in-line 
tips['sex-smoker'] = tips.apply(lambda row: concat(row['sex'], row['smoker']), axis=1)
```

```python
tips['sex-smoker'].unique()
```

```python
# this will add a best fit line ... regression
# total_bill vs tip by day
sns.lmplot(data=tips, x="total_bill", y="tip", hue="sex-smoker")
```

```python
su_map = folium.Map(location=(43.0387, -76.1337), zoom_start=20)
su_map
```

```python
import pandas as pd
classesdf = pd.read_csv("https://raw.githubusercontent.com/mafudge/datasets/master/delimited/class-schedule.csv")
classesdf
```

```python
folium.Marker(location=(43.0387, -76.1337), popup="hi", tooltip="low").add_to(su_map)
```

```python
su_map
```

```python
for i, row in classesdf.iterrows():
    loc = (row['Lat'], row['Lon'])
    folium.Marker(location=loc, popup=row['Course']).add_to(su_map)
```
