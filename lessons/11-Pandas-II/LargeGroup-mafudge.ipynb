```python
players = pd.read_csv("https://raw.githubusercontent.com/mafudge/datasets/refs/heads/master/delimited/bbplayers.csv")
```

```python
teams = pd.read_csv("https://raw.githubusercontent.com/mafudge/datasets/refs/heads/master/delimited/bbteams.csv")
```

```python
players
```

```python
teams
```

```python
player_teams = pd.merge(left=players,
                        right=teams,
                        how="inner",
                        left_on="player_team_id",
                        right_on="team_id")
```

```python
all_players_teams = pd.merge(left=players,
                            right=teams,
                            how="left",
                            left_on="player_team_id",
                            right_on="team_id")
all_players_teams
```

```python
def hall_of_fame_candidate(points:int) -> str:
    if points > 30000:
        return "high"
    if points > 20000:
        return "medium"
    
    return "low"

assert hall_of_fame_candidate(50000) == "high"
assert hall_of_fame_candidate(1000) == "low"
assert hall_of_fame_candidate(20001) == "medium"
```

```python
players
```

```python
players['hof'] = players.apply(lambda row: hall_of_fame_candidate(row['career_pts']), axis=1)
```
