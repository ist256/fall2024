```python
number_of_100s = 0
grade_count = 0
with open("grades.txt", "r") as f:
    for line in f.readlines():
        grade = int(line)
        if grade == 100:
            number_of_100s += 1
        grade_count += 1
print(f"The file grades.txt has {grade_count} grades and {number_of_100s} perfect scores")
```
